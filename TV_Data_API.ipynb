```python
import pandas as pd
import numpy as np
import json
import requests
```

```python
#Create a total results list
total_results =[]


#This loop will iterate through all 36 pages of the TMDB TV API. 
#Each iteration it will extract the JSON object from each page and append it into total results
for page_num in range(1, 37):
    url = 'https://api.themoviedb.org/3/tv/top_rated?api_key=bd54255760b7c9a9eb0d294603c9610e&language=en-US&page='+str(page_num)
    print("Downloading", url)
    response = requests.get(url)
    data = response.json()
    total_results = total_results + data['results']
    
print("We have", len(total_results), "total results")
```

```python
df = pd.DataFrame(total_results)
```
